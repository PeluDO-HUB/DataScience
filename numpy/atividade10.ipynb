1 – Crie dois dataFrames (Tabelas):
•O Primeiro: Filtre na coluna “grape_type” somente os valores “Wine”, filtre na coluna “location_type” o valor “Brazil_State” e por fim, a coluna “category” o valor “Sales”;
•O Segundo: Filtre na coluna “grape_type” somente os valores “Table” e a coluna “category” o valor “Sales”.
No fim, use describe().T, qual das duas possuí a maior média na coluna “value_usd”?

```python
import pandas as pd


df01 =pd.read_excel("vinhos_exercicio.xlsx")
df02 = pd.read_excel("vinhos_exercicio.xlsx")
dfOrigem=pd.read_excel("vinhos_exercicio.xlsx")

df01 = df01[
    (df01["grape_type"] == "Wine") &
    (df01["location_type"] == "Brazil_State") &
    (df01["category"] == "Sales")
]
print(df01)

df02 = df02[
    (df02["grape_type"]=="Table") &
    (df02["category"]=="Sales")
]
print(df02)

print(df01.describe().loc["mean"].T.compare(df02.describe().loc["mean"].T))

mediadf01=df01["value_usd"].mean()
mediadf02=df02["value_usd"].mean()
if mediadf01>mediadf02:
    print(f"A média do DF01: {mediadf01:.2f} é maior que do DF02: {mediadf02:.2f} e a diferença é de: {(mediadf01-mediadf02):.2f}")
else:
    print(f"A média do DF02: {mediadf02:.2f} é maior que do DF01: {mediadf01:.2f} e a diferença é de: {(mediadf02-mediadf01):.2f}")
```
