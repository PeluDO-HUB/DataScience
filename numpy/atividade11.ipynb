1 – Crie pelo menos 20 valores vazios pelas colunas em um df3 e dupliqueo. Mostre o shape e o describe()

```python
import pandas as pd
import numpy as np
dataOrigem=pd.read_excel("vinhos_exercicio.xlsx")
df3=dataOrigem.copy()



posicoes = set()
while len(posicoes) < 20:
    i = np.random.randint(0, df3.shape[0])
    j = np.random.randint(0, df3.shape[1])
    posicoes.add((i, j))

for i, j in posicoes:
    df3.iat[i, j] = np.nan
print(df3.isnull().sum().sum())

df3_copia= pd.concat([df3,df3])

print(df3_copia.shape)
print(df3_copia.describe().T)
```

2 – Contabilize os valores nulos em cada coluna e total, faça dois df’s (df4 e df5), um tirando as duplicatas o outro os valores vazios. Mostre o shape e o describe() de cada um.

```python
print("-"*22+"DF3 Original"+"-"*22)
print(df3.isnull().sum())
print(f"Total: {df3.isnull().sum().sum()}")
print("-"*22+"DF3 Duplicado"+"-"*22)
print(df3_copia.isnull().sum())
print(f"Total: {df3_copia.isnull().sum().sum()}")

df4 = df3_copia.copy()
df5 = df3_copia.copy()

df4 = df4.drop_duplicates()
df5 = df5.dropna()

print("-"*22+"DF4 retirando as duplicadas"+"-"*22)
print(df4.shape)
print(df4.describe().T)
print("-"*22+"DF5 retirando os vazios"+"-"*22)
print(df5.shape)
print(df5.describe().T)
```

3 – Crie o dfFinal, a partir do df3, realizando a retirada dos duplicados, monstrando o shape do antes e depois. Por fim subistitua os vazios de cada coluna (para volume_hl e value_usd, substitua usando a mediana median()).

```python
dfFinal= df3.copy()
print(df3_copia.shape)
dfFinal=dfFinal.drop_duplicates()
print(dfFinal.shape)

dfFinal["volume_hl"]=dfFinal["volume_hl"].fillna(dfFinal["volume_hl"].median())
dfFinal["value_usd"]=dfFinal["value_usd"].fillna(dfFinal["value_usd"].median())
```

4 – De forma formatada print o antes e depois do dfFinal e do df3, mostrando o shape o describe().T. Compare-os com o df original.

```python
print("-"*22+"DF3 Original"+"-"*22)
print(df3_copia)
print(df3_copia.shape)
print(df3_copia.describe().T)
print("-"*22+"DF Final"+"-"*22)
print(dfFinal)
print(dfFinal.shape)
print(dfFinal.describe().T)
print("-"*22+"Comparando DF Final com Origem"+"-"*22)
print(dfFinal.describe().T.compare(dataOrigem.describe().T))
print("-"*22+"Comparando DF 3 com Origem"+"-"*22)
print(df3_copia.describe().T.compare(dataOrigem.describe().T))
```
