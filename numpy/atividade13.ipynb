1.Construa uma tabela contendo frequência absoluta, relativa (%) e acumulada da coluna grape_type.



```python
import pandas as pd

df = pd.read_excel("vinhos_exercicio.xlsx")


freq_abs = df["grape_type"].value_counts()
freq_rel = df["grape_type"].value_counts(normalize=True) * 100
freq_acum = freq_abs.cumsum()

tabela_grape = pd.DataFrame({
    "Freq. Absoluta": freq_abs,
    "Freq. Relativa (%)": freq_rel.round(2),
    "Freq. Acumulada": freq_acum
})

print("\n"+"-"*20+"Por tipo de Uva (grape_type)"+"-"*20)
print(tabela_grape)

```

2.Use pd.qcut para dividir value_usd em 4 classes iguais (quartis) e gere uma tabela com frequência absoluta e relativa de cada quartil. Use o parâmetro q=4.

```python
quartis = pd.qcut(df["value_usd"], q=4)

freq_abs_q = quartis.value_counts().sort_index()
freq_rel_q = quartis.value_counts(normalize=True).sort_index() * 100

tabela_quartis = pd.DataFrame({
    "Freq. Absoluta": freq_abs_q,
    "Freq. Relativa (%)": freq_rel_q.round(2)
})

print("\n"+ "-"*20+"Quartis Valor (value_usd)"+ "-"*20)
print(tabela_quartis)


```

3.Construa uma tabela contendo frequência absoluta, relativa (%) e acumulada da coluna year, usando as seguintes faixas (2014-2018, 2019-2022, 2023-2027).

```python
limite = [2014, 2018, 2022, 2027]
margem = ["2014-2018", "2019-2022", "2023-2027"]

periodo = pd.cut(df["year"], bins=limite, labels=margem, include_lowest=True)

freq_abs_ano = periodo.value_counts().sort_index()
freq_rel_ano = periodo.value_counts(normalize=True).sort_index() * 100
freq_acum_ano = freq_abs_ano.cumsum()

tabela_ano = pd.DataFrame({
    "Freq. Absoluta": freq_abs_ano,
    "Freq. Relativa (%)": freq_rel_ano.round(2),
    "Freq. Acumulada": freq_acum_ano
})

print("\n"+ "-"*20+"Por Periodo"+ "-"*20)
print(tabela_ano)
```
