  1. Tratamento de Dados (Python)  
Utilizando Python no VS Code:

Carregar a base de dados

Identificar e tratar valores nulos

Remover duplicidades

Criar novas colunas (ex: preço médio)

Garantir a qualidade dos dados

```python
import pandas as pd
import numpy as np

dfOrigem = pd.read_excel("vinhos_desafio.xlsx")
print(f"-"*22+"DF Origem Nulos"+"-"*22)
print(dfOrigem.isnull().sum())
print(f"Total: {dfOrigem.isnull().sum().sum()}")

dfTratado = dfOrigem.copy()
dfTratado["volume_hl"] = dfTratado["volume_hl"].fillna(dfTratado["volume_hl"].mean())
dfTratado["valor_usd"] = dfTratado["valor_usd"].fillna(dfTratado["valor_usd"].mean())

print(f"-"*22+"DF Tratado Nulos sendo substituidos pela media da coluna"+"-"*22)
print(dfTratado.isnull().sum())
print(f"Total: {dfTratado.isnull().sum().sum()}")

print(f"-"*22+"Tratando as colunas duplicadas e comparando o Origem com o Tratado"+"-"*22)
print(dfTratado.shape)
dfTratado = dfTratado.drop_duplicates()
print(dfOrigem.shape,dfTratado.shape)

#Evitar o resultado inf
dfTratado["volume_hl"] = dfTratado["volume_hl"].replace(0, np.nan)
dfTratado["preco_medio"]= dfTratado["valor_usd"]/dfTratado["volume_hl"]

dfTratado



```

2. Análise Exploratória

Ainda em Python:

Identificar padrões nos dados

Comparar vendas, importação e exportação

Analisar evolução ao longo dos anos

Avaliar diferenças entre tipos de vinho

```python

print("-"*22 +"ESTATÍSTICAS"+"-"*22)
print(dfTratado.describe().T)

print("-"*22+"Correlação"+"-"*22)
print(dfTratado.corr(numeric_only=True))


if "categoria" in dfTratado.columns:
    print("-"*22+"Por Categoria"+"-"*22)
    print(dfTratado.groupby("categoria")[["volume_hl", "valor_usd"]].sum().round(2))

    print("\n"+"Preço médio por tipo:")
    print(dfTratado.groupby("categoria")["preco_medio"].mean().round(2))


if "ano" in dfTratado.columns:
    print("-"*22+"Evolução Anual"+"-"*22)
    print(dfTratado.groupby("ano")[["volume_hl", "valor_usd"]].sum().round(2))

    print("\n"+"Preço médio por ano:")
    print(dfTratado.groupby("ano")["preco_medio"].mean().round(2))


if "tipo_uva" in dfTratado.columns:
    print("-"*22 +"Tipos de Uva"+"-"*22)

    print("\n"+"Preço médio por tipo de vinho:")
    print(dfTratado.groupby("tipo_uva")["preco_medio"].mean().sort_values(ascending=False).round(2))

    print("\n"+"Volume total por tipo de vinho:")
    print(dfTratado.groupby("tipo_uva")["volume_hl"].sum().round(2))

# ==============================
# 7. Rankings
# ==============================
print("\n"+"-"*22+"TOP Preços"+"-"*22)
print(dfTratado.sort_values(by="preco_medio", ascending=False).head().round(2))

print("\n"+"-"*22+"MAIOR VOLUME "+"-"*22)
print(dfTratado.sort_values(by="volume_hl", ascending=False).head().round(2))

```
