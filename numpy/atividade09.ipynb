3 – Adicione pelo menos 20 valores vazios espalhados pelas colunas em um dataFrame cópia, após isso use o .isnull() com .sum() para ver como ficou a dispersão de vazios

```python
data = {
    'Produto': [f'Item {i}' for i in range(1, 11)],
    'Vendas': [100, 150, 200, 120, 90, 300, 250, 180, 210, 140],
    'Estoque': [10, 20, 15, 30, 5, 40, 25, 12, 18, 22],
    'Preço': [15.5, 20.0, 10.5, 45.0, 30.0, 12.0, 18.5, 22.0, 14.0, 50.0]
}

df = pd.DataFrame(data)
df_copia = df.copy()

colunas = df_copia.columns.tolist()
num_linhas = len(df_copia)

for _ in range(20):
    linha_aleatoria = random.randint(0, num_linhas - 1)
    coluna_aleatoria = random.choice(colunas)
    df_copia.loc[linha_aleatoria, coluna_aleatoria] = None

print("Relatório de valores ausentes por coluna:")
print(df_copia.isnull().sum())
```

4 – Nesse dataFrame cópia utilize o df.describe().T e o compare com o do dataFrame orginal


```python
print(df_copia.describe().T)
print(df_copia.describe().T.compare(df.describe().T,keep_equal=True))
```
