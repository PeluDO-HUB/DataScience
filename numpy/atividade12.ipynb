1.Sorteie 5 amostras de 5 % cada uma; registre médias de volume_hl. 

```python
import pandas as pd
df = pd.read_excel("vinhos_exercicio.xlsx")

media_populacao =  df["volume_hl"].mean()
amostra = df.sample(frac=0.05,random_state=5)
media_amostra = amostra["volume_hl"].mean()
print(round(media_populacao,2),round(media_amostra,2))
```


2.Compare médias de amostras e a média da população.

3.Explique em 1 frase por que amostras podem variar.

elas podem variar por conta do random_state especificado dentro da amostra como escolhidos aleatoriamentet

```python
if media_amostra > media_populacao:
    print(f"A amostra tem maior media que a população, diferença de : {(media_amostra-media_populacao):.2f}")
else:
    print(f"A população tem maior media que a amostra , diferença de : {(media_populacao-media_amostra):.2f}")
```
