```python
data= {"Nome":["Ana","Bruno","Caio"],"Idade":[25,30,20]}

dataf= pd.DataFrame(data)

#como filtrar um dado, & para "e" e | para "ou"
filtro=dataf[(dataf["Idade"]>22) & (dataf["Nome"]!="Ana")]
print(filtro)

```
